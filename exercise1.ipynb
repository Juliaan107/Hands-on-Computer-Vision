```python
# @title imports
import matplotlib.pyplot as plt
import numpy as np
import requests
from PIL import Image
from transformers import pipeline
```

```python
# @title Rellena estos campos
nombre = "Julian" # @param {type:"string"}
codigo = "2232451" # @param {type:"string"}
```

```python
# @title cargar imagen
url1 = "https://assets.khelnow.com/news/uploads/2025/06/Lionel-Messi-Argentina-GettyImages-2177987340-1-scaled.jpg"

url2 = "https://editorial01.shutterstock.com/preview/13647945nq/9cb34c8f/Shutterstock_13647945nq.jpg"
# puedes poner urls de internet!

img1 = Image.open(requests.get(url1, stream=True).raw)
img2 = Image.open(requests.get(url2, stream=True).raw)


plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.axis("off")

plt.show()
```

```python
# @title clasificar imagen
classifier = pipeline('image-classification')
result1 = classifier(url1)
result2 = classifier(url2)

print(f"\n **La imagen es un: {result1[0]['label']}** ")
print(f"\n **La imagen es un: {result2[0]['label']}** ")
```
